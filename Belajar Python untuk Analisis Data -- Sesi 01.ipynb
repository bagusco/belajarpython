# Membuat Variabel Baru pada Suatu Dataframe

Ada kalanya dalam suatu analisis kita perlu membuat variabel-variabel baru berdasarkan variabel yang sudah ada.  Misalnya penjumlahan dari tiga variabel menjadi suatu variabel total.  Atau, menghitung rasio dari dua variabel.  Atau, mengganti satuan data dari meter menjadi centimeter.

Pada kasus data RUMAH, kita memiliki kolom "harga" yang satuannya adalah juta rupiah.  Seandainya kita ingin membuat kolom baru yang berisi harga dalam satuan rupiah, maka variabel baru tersebut tidak lain merupakan perkalian dari variabel harga dengan konstanta 1000000.  Berikut ini caranya.

```python
#membuat kolom baru dengan nama 'harga_rupiah' yang merupakan perkalian harga dengan 1000000
rumah['harga_rupiah'] = rumah['harga'] * 1000000
```

```python
#menampilkan hasil penambahan kolom baru
rumah.head()
```

```python
#membuat kolom baru dengan nama 'rasio_bangunan' yang merupakan rasio antara 'luasbangunan' dengan 'luastanah'
rumah['rasio_bangunan'] = rumah['luasbangunan'] / rumah['luastanah']
rumah.head()
```
